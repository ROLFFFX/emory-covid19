```python
import csv
import json
import pandas as pd
import numpy as np
from pprint import pprint
from collections import defaultdict, Counter
from dateutil.parser import *

SMA_WINDOW = 7

def get_lineitem(d):
    cases = None
    deaths = None
    caseRate = None
    mortality = None
    racePopulation = None
    populationTot = None
    percentPop = None

        
    try:
        cases = int(d["cases"])
        deaths = int(d["deaths"])
        caseRate = float(d["caserate"])
        mortality = float(d["deathrate"])
        racePopulation = float(d["racePopulation"])
        populationTot = float(d["populationTot"])
        percentPop = int(d["percentPop"])

        
    except ValueError:
        cases = 0
        deaths = 0
        caseRate = 0
        mortality = 0
        racePopulation = 0
        populationTot = 0
        percentPop = 0

    

    return {
            "cases": cases,
            "deaths": deaths,
            "caseRate": caseRate,
            "mortality": mortality,
            "racePopulation": racePopulation,
            "populationTot": populationTot,
            "percentPop": percentPop}


def linechart(fn="staticracedata.csv"):
    data = defaultdict(list)
    with open(fn, "r") as fp:
        reader = csv.reader(fp)
        header = next(reader)
        for row in reader:
            
                d = {k:v.strip() for k, v in zip(header, row)}
                item = get_lineitem(d)
                k = ""
                k = d["state"]

                data[k].append(item)
            
            
        

    item0 = {"cases": 0,
            "deaths": 0,
            "caseRate": 0,
            "mortality": 0,
            "racePopulation": 0,
            "populationTot": 0,
            "percentPop": 0}
    item1 = {"cases": 0,
            "deaths": 0,
            "caseRate": 0,
            "mortality": 0,
            "racePopulation": 0,
            "populationTot": 0,
            "percentPop": 0}
    
    itemNA = {"cases": 0,
            "deaths": 0,
            "caseRate": 0,
            "mortality": 0,
            "racePopulation": 0,
            "populationTot": 0,
            "percentPop": 0}
    data["99999"] = [itemNA]

    data["_"] = [item0, item1]

    output = defaultdict(dict)
    for k, v in data.items():
        if k == "_nation":
            continue
        output[k[:2]][k] = v

    for k, v in output.items():
        v["_"] = data["_"]
        v["99999"] = data["99999"]
        with open(f"../staticracedata{k}.json", "w") as fp:
            json.dump(v, fp, indent=2)


if __name__=="__main__":

    linechart()

```
